```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data = pd.read_csv('Housing.csv')
data.head()
```

```python
# Scatter plot (price vs. Area)
plt.figure(figsize=(7,7))
sns.scatterplot(x = 'area', y='price', data=data)
plt.title('PRICE VS AREA')
plt.xlabel('Area (sq.ft)')
plt.ylabel('Price ($)')
plt.show()
```

```python
# BAR CHART

plt.figure(figsize=(7,7))
data['stories'].value_counts().plot(kind='bar')
plt.title('Number of properties as perlocation')
plt.xlabel('stories')
plt.ylabel('Properties')
plt.show()
```

```python
# 3. Histogram (price Distribution)
plt.figure(figsize=(8,6))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price ($)')
plt.show()
```

```python
# 4 box plot
plt.figure(figsize=(8,6))
sns.boxplot(x='bedrooms', y='price', data= data)
plt.title('price per Number of bedrooms')
plt.xlabel('Number of Beadrooms')
plt.ylabel('price ($)')
plt.show()
```

```python
price = data['price'].value_counts()  # Correctly count the occurrences of each price

plt.figure(figsize=(8,6))
plt.pie(price, labels=price.index, autopct='%1.1f%%', startangle=90)  # Plot pie chart with correct data
plt.title('Price Distribution')  # Correct the title capitalization
plt.show()
```

```python
# 6. Heatmap (Bedrooms vs. Bathrooms)
plt.figure(figsize=(8,6))
heatmap_data = pd.crosstab(data['bedrooms'], data['bathrooms'])
sns.heatmap(heatmap_data, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Bedrooms vs. Bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('Bedrooms')
plt.show()
```
